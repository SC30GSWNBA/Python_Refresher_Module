```python
bins=[16,18,20,22,24]
labels1=["16-18","18-20","20-22","22-24"]
weather_tidy["max_cat"]=pd.cut(weather_tidy["max"],bins=bins,labels=labels1)
weather_tidy
```

```python
bins=[16,18,20,22,24]
labels1=["low","medium","high","very high"]
weather_tidy["max_cat_new labels"]=pd.cut(weather_tidy["max"],bins=bins,labels=labels1)
weather_tidy
```

```python
bins=[16,18,22,24]
labels1=["low","medium","very high"]
weather_tidy["max_cat_3_labels"]=pd.cut(weather_tidy["max"],bins=bins,labels=labels1)
weather_tidy
```

```python
bins=[16,18,22,24]
labels1=["low","medium","high"]
weather_tidy["max_cat_3_labels1"]=pd.cut(weather_tidy["max"],bins=bins,labels=labels1)
weather_tidy
```
