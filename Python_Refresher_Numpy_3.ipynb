**Matrix multiplication**

There are 3 different types of multiplication possible in Numpy

> Scaler multiplication - Eg ((1 * 2) or (-5 * 9))

> Element-wise multiplication - Eg ([1,2,3], [4,5,6])

> Matrix Multiplication (arrays with same/different dimension and same/different shape) - Eg ([1,2,3], [[1,2,3], [4,5,6], [7,8,9]]

```python
# IMPORTANT - a*b is possible only when both the arrays are of the same dimension and same size

# Element wise multiplication in numpy arrays

print(a)
print(b)
a*b
```

```python
# IMPORTANT

# Matrix multiplication in numpy arrays - for arrays of different size
# Prerequisite - A with shape(R1, C1) and B with shape(R2, C2), C1 == R2 is a must

np.dot(a,b)
```

# Broadcasting Rules

### For each dimension ( going from right side)
#### 1. The size of each dimension should be same OR
#### 2. The size of one dimension should be 1


##### Rule 1 : If two array differ in the number of dimensions, the shape of one with fewer dimensions is padded with ones on its leading( Left Side).
##### Rule 2 : If the shape of two arrays doesnt match in any dimensions, the array with shape equal to 1 is stretched to match the other shape.
##### Rule 3 : If in any dimesion the sizes disagree and neither equal to 1 , then Error is raised.
