**Aggregate functions on DataFrameGroupBy object**

```python
df_agg=df.groupby(["director_name"])[["year","title"]].aggregate({"year":["min","max"],"title":["count"]})
```

**Converging tuples columns into a single column name**

```python
df_agg.columns=["_".join(col) for col in df_agg.columns]
df_agg
```

**Reseting index for a dataframe**

```python
df_agg["years_active"]=df_agg["year_max"]-df_agg["year_min"]
df_agg
```

**Filter and Apply methods in a pandas dataframe**

```python
df_filter=df.groupby(["director_name"]).filter(lambda x:x['budget'].max()>=100)
```

```python
def sub_avg(x):
  x=x-x.mean()
  return x
df.groupby(["director_name"])['budget'].transform(sub_avg)
```

```python
def func(x):
  x["risky"]=(x["budget"]-x["revenue"].mean()) >= 0
  return x
df_risky=df.groupby(['director_name']).apply(func)
```

```python
df_risky
```

```python
df_risky.loc[df_risky["risky"]]
```
